```python
from keras import models
import numpy as np
import os
import pandas



file_path=os.path.dirname('__file__') #현재 파일 경로
model_path=os.path.join(file_path,'iris_model.h5') #모델 파일 경로
model=models.load_model(model_path) #모델 로드



# 테스트 데이터 처리 
csv_pile_path=os.path.join(file_path,'iris.csv')
data=pandas.read_csv(csv_pile_path,sep=",") #데이터 로드

feature_data=data[["Sepal.Length","Sepal.Width","Petal.Length","Petal.Width"]] #특징 데이터만 추출
feature_data=np.array(feature_data)#데이터 벡터화



# 예측
predictions=model.predict(feature_data) 

species_decoding=['setosa','versicolor','virginica'] #인덱스 문자열
for i in range(predictions.shape[0]):
    species=np.argmax(predictions[i]) #가장 확률 높은 종 인덱스 리턴
    print(species_decoding[species])
```
